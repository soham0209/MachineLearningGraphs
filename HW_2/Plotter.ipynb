```python
import numpy as np
file = 'recall_ppr.txt'
recall = np.loadtxt(file)
print(recall)
```

```python
recall_1 = recall[recall[:, 0] == 0.95][:,1:]
recall_2 = recall[recall[:, 0] == 0.98][:,1:]
print(recall_1)
print(recall_2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
plt.plot(recall_2[:, 0], recall_2[:, 1])
plt.suptitle('alpha = 0.98')
plt.xlabel('k')
plt.ylabel('mean recall@top-100')
plt.savefig('alpha_2.png')
plt.show()
```
